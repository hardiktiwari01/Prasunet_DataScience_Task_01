```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the data from the provided CSV file
file_path = r'C:\Users\HARDIK TIWARI\Desktop\datasets\worldpopulationdata.csv'
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset to understand its structure
print(data.head())
```

```python
# Data Cleaning: Handle missing values
# Drop rows where the population data for 2022 is missing
data_cleaned = data.dropna(subset=['2022'])
```

```python
# Additional Data Cleaning: Convert the population column to numeric if needed
# This ensures that all population data is treated as numbers
data_cleaned['2022'] = pd.to_numeric(data_cleaned['2022'], errors='coerce')

# Handle missing or NaN values after conversion
data_cleaned = data_cleaned.dropna(subset=['2022'])

# Summary statistics for population in 2022
print("\nSummary statistics for the population in 2022:")
print(data_cleaned['2022'].describe())
```

```python
# Visualization 1: Bar Chart for Population Distribution by Country in 2022
plt.figure(figsize=(16, 8))
sns.barplot(x='Country Name', y='2022', data=data_cleaned, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Population in 2022')
plt.title('Population Distribution by Country in 2022')
plt.show()

```

```python
# Visualization 2: Histogram of Population Distribution in 2022
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['2022'], bins=20, kde=True, color='blue')
plt.xlabel('Population in 2022')
plt.ylabel('Number of Countries')
plt.title('Histogram of Population Distribution in 2022')
plt.show()
```

```python
# Visualization 3: Boxplot to Identify Outliers in Population Data
plt.figure(figsize=(10, 6))
sns.boxplot(x=data_cleaned['2022'])
plt.xlabel('Population in 2022')
plt.title('Boxplot of Population Distribution in 2022')
plt.show()
```
